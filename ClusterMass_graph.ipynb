```python
import matplotlib.pyplot as plt
import pandas as pd

data_fl= pd.read_csv("alpha_IC2714_func_flat.txt", delimiter='\t')
#data_gauss= pd.read_csv("alpha_IC2714_func_gauss.txt", delimiter='\t')
#data_gr= pd.read_csv("alpha_IC2714_func_gr.txt", delimiter='\t')
data_corrected = pd.read_csv("alpha_IC2714_func_flat_corrected.txt", delimiter='\t')
data_corrected_high = pd.read_csv("alpha_IC2714_func_flat_corrected_high.txt", delimiter='\t')
data_corrected_low = pd.read_csv("alpha_IC2714_func_flat_corrected_low.txt", delimiter='\t')
data_same_4 = pd.read_csv("alpha_IC2714_same_4.txt", delimiter='\t')
data_same_4_high = pd.read_csv("alpha_IC2714_same_4_high.txt", delimiter='\t')
data_same_4_low = pd.read_csv("alpha_IC2714_same_4_low.txt", delimiter='\t')
data_same_2 = pd.read_csv("alpha_IC2714_same_2.txt", delimiter='\t')
```

```python
fig, ax = plt.subplots(figsize=(16,14 )) 
ax.plot(data_fl['ALPHA'], data_fl['RATIO'], c = 'black', lw = 2, label =' ') 
#ax.plot(data_gauss['ALPHA'], data_gauss['RATIO'], c = 'red', lw = 2, label ='    0.23') 
#ax.plot(data_gr['ALPHA'], data_gr['RATIO'], c = 'blue', lw = 2, label ='    0.60')
ax.plot(data_corrected['ALPHA'], data_corrected['RATIO'], c = 'blue', lw = 2, label = '     ')
ax.plot(data_corrected_high['ALPHA'], data_corrected_high['RATIO'], c = 'blue', lw = 1.5, ls='--',label ='')
ax.plot(data_corrected_low['ALPHA'], data_corrected_low['RATIO'], c = 'blue', lw = 1.5, ls='-.',label ='')

ax.plot(data_same_4['ALPHA'], data_same_4['RATIO'], c = 'red', lw = 2, label = '     ')
ax.plot(data_same_4_high['ALPHA'], data_same_4_high['RATIO'], c = 'red', lw = 1.5, ls='--',label ='')
ax.plot(data_same_4_low['ALPHA'], data_same_4_low['RATIO'], c = 'red', lw = 1.5, ls='-.',label ='')

ax.plot(data_same_2['ALPHA'], data_same_2['RATIO'], c = 'green', lw = 2, label = '  ')

ax.errorbar(data_fl['ALPHA'], data_fl['RATIO'], yerr = data_fl['RATIO SD'], c = 'black', label = None)
locs, labels = plt.yticks()
locs, labels = plt.xticks()
ax.set_xlim(0)
ax.set_ylim(1)
ax.set_title("IC2714", size = 40)
ax.grid(c='#aaaaaa', ls='--')
plt.legend(fontsize = 20)
plt.xlabel(' , \u03B1', size = 30)
plt.ylabel("   \n  /   ".expandtabs(), size = 30)
ax.tick_params(axis='both', which='major', labelsize=25)
plt.savefig("alphas_IC2714.png", dpi =200)
```
