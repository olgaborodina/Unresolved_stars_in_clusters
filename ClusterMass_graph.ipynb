```python
import matplotlib.pyplot as plt
import pandas as pd
from pathlib import Path
```

```python
cluster = 'IC2714'
cluster_dir = Path(f'../{cluster}/')
cluster_h = Path(f'../{cluster}/high/')
cluster_l = Path(f'../{cluster}/low/')
```

```python
data_gauss = pd.read_csv(cluster_dir / f"alpha_{cluster}_func_gauss_corrected.txt", delimiter='\t')
data_gauss_high = pd.read_csv(cluster_h / f"alpha_{cluster}_func_gauss_corrected_high.txt", delimiter='\t')
data_gauss_low = pd.read_csv(cluster_l / f"alpha_{cluster}_func_gauss_corrected_low.txt", delimiter='\t')

data_gr = pd.read_csv(cluster_dir / f"alpha_{cluster}_func_gr_corrected.txt", delimiter='\t')
data_gr_high = pd.read_csv(cluster_h / f"alpha_{cluster}_func_gr_corrected_high.txt", delimiter='\t')
data_gr_low = pd.read_csv(cluster_l / f"alpha_{cluster}_func_gr_corrected_low.txt", delimiter='\t')

data_fl = pd.read_csv(cluster_dir / f"alpha_{cluster}_func_flat_corrected.txt", delimiter='\t')
data_fl_high = pd.read_csv(cluster_h / f"alpha_{cluster}_func_flat_corrected_high.txt", delimiter='\t')
data_fl_low = pd.read_csv(cluster_l / f"alpha_{cluster}_func_flat_corrected_low.txt", delimiter='\t')

data_same_4 = pd.read_csv(cluster_dir / f"alpha_{cluster}_same_4.txt", delimiter='\t')
data_same_4_high = pd.read_csv(cluster_h / f"alpha_{cluster}_same_4_high.txt", delimiter='\t')
data_same_4_low = pd.read_csv(cluster_l / f"alpha_{cluster}_same_4_low.txt", delimiter='\t')

data_same_2 = pd.read_csv(cluster_dir / f"alpha_{cluster}_same_2.txt", delimiter='\t')
data_same_2_high = pd.read_csv(cluster_h / f"alpha_{cluster}_same_2_high.txt", delimiter='\t')
data_same_2_low = pd.read_csv(cluster_l / f"alpha_{cluster}_same_2_low.txt", delimiter='\t')
```

```python
plt.rcParams.update({'font.size': 15})
plt.rc('text', usetex=True)
plt.rc('font', family='serif')
plt.rc('text.latex', preamble=r'\usepackage[utf8]{inputenc}')
plt.rc('text.latex', preamble=r'\usepackage[russian]{babel}')
```

```python
fig, ax = plt.subplots(figsize=(9,10)) 

ax.plot(data_gauss['ALPHA'], data_gauss['RATIO'], c = 'purple', lw = 2, label ='    0.23') 
ax.plot(data_gauss_high['ALPHA'], data_gauss_high['RATIO'], c = 'purple', lw = 1.5, ls='--',label ='')
ax.plot(data_gauss_low['ALPHA'], data_gauss_low['RATIO'], c = 'purple', lw = 1.5, ls='-.',label ='')

ax.plot(data_gr['ALPHA'], data_gr['RATIO'], c = 'blue', lw = 2, label ='    0.60')
ax.plot(data_gr_high['ALPHA'], data_gr_high['RATIO'], c = 'blue', lw = 1.5, ls='--',label ='')
ax.plot(data_gr_low['ALPHA'], data_gr_low['RATIO'], c = 'blue', lw = 1.5, ls='-.',label ='')

ax.plot(data_fl['ALPHA'], data_fl['RATIO'], c = 'black', lw = 2, label = ' ')
ax.plot(data_fl_high['ALPHA'], data_fl_high['RATIO'], c = 'black', lw = 1.5, ls='--',label ='')
ax.plot(data_fl_low['ALPHA'], data_fl_low['RATIO'], c = 'black', lw = 1.5, ls='-.',label ='')

ax.plot(data_same_4['ALPHA'], data_same_4['RATIO'], c = 'red', lw = 2, label = '     ')
ax.plot(data_same_4_high['ALPHA'], data_same_4_high['RATIO'], c = 'red', lw = 1.5, ls='--',label ='')
ax.plot(data_same_4_low['ALPHA'], data_same_4_low['RATIO'], c = 'red', lw = 1.5, ls='-.',label ='')

ax.plot(data_same_2['ALPHA'], data_same_2['RATIO'], c = 'green', lw = 2, label = '  ')
ax.plot(data_same_2_high['ALPHA'], data_same_2_high['RATIO'], c = 'green', lw = 1.5, ls='--',label ='')
ax.plot(data_same_2_low['ALPHA'], data_same_2_low['RATIO'], c = 'green', lw = 1.5, ls='-.',label ='')

ax.errorbar(data_fl['ALPHA'], data_fl['RATIO'], yerr = data_fl['RATIO SD'], c = 'black', label = None)
ax.errorbar(data_gr['ALPHA'], data_gr['RATIO'], yerr = data_fl['RATIO SD'], c = 'blue', label = None)
ax.errorbar(data_gauss['ALPHA'], data_gauss['RATIO'], yerr = data_fl['RATIO SD'], c = 'purple', label = None)

locs, labels = plt.yticks()
locs, labels = plt.xticks()
ax.set_xlim(0)
ax.set_ylim(1,2.3)
ax.set_title(f"{cluster}", size = 40)
ax.grid(c='#aaaaaa', ls='--')
plt.legend(fontsize = 17)
plt.xlabel(r' , $\alpha$', size = 30)
plt.ylabel(r"   \\  /   ".expandtabs(), size = 30)
ax.tick_params(axis='both', which='major', labelsize=25)
plt.savefig(f"alphas_{cluster}.png", bbox_inches='tight')
```
